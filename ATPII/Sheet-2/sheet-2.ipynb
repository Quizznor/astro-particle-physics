# 5 Tscherenkow-Strahlung


Die Tscherenkow-Strahlung (oder auch Wawilow-Tscherenkow-Strahlung), die von hochenergetischen Teilchen wie z.B. Atomkernen der kosmischen Strahlung in geeigneten Medien produziert wird, wurde beim HEAO-3-Satelliten der NASA genutzt, um die chemische Zusammensetzung der ionisierten Elemente kosmischer Strahlung zu bestimmen. Aerogel, ein sehr leichter und hochporöser Feststoff mit einem Brechungsindex von $n = 1.053$, gehort zu den Materialien, die Tscherenkow-Strahlung erzeugen und wurde beim HEAO-3-Satelliten dafur eingesetzt.

#### Zeigen Sie, dass ein Atomkern mindestens eine Energie von etwas mehr als dreimal seiner Ruheenergie benotigt, um im HEAO-3-Satelliten detektiert zu werden.

##### Lösung Aufgabe 5:

Cherenkov Effekt tritt auf für $\beta\geq\frac{c}{c_n} = \frac{1}{n}$, bzw. für Teilchen mit der Energie $$ E = \frac{m_0 c²}{\sqrt{1-\beta²}}. $$ 
Daraus folgt für die Mindestenergie zur Detektion durch HEAO-3 $$ E_\text{min.} = \frac{m_0 c²}{\sqrt{1-\frac{1}{n²}}}. $$

```python
import math

n_gel = 1.053
c = 299792        # km/s
c_gel = c/n_gel   # m/s

E_min = 1/math.sqrt(1-1/n_gel**2)

print(f"\nLichtgeschwindigkeit in Aerogel: c_gel = {c_gel:.3e} km/s")
print(f"Minimalenergie for Cherenkov-Eff.: E_min = {E_min:.2f} 1/mc²\n")
```

# 6 Auflösungseffekte beim gemessenen Teilchenfluss

Ein Satellitenexperiment misst den Fluss von Protonen oberhalb der Erdatmosphare. Dabei kann die Energie jedes der gemessenen Protonen nur mit einer gewissen Messgenauigkeit bestimmt werden. Der Logarithmus des Verhältnisses zwischen der wahren und der rekonstruierten Energie ist normalverteilt mit einer Varianz von $σ^2$. Das Verhältnis zwischen wahrer und rekonstruierter Energie ist demnach log-normal verteilt.


#### Wie groß ist der tatsächliche Fluss, $Φ(E_{real})$, der Protonen, wenn der Satellit einen Fluss gemäß eines Potenzgesetzes von $Φ(E_{rec}) = Φ_0 E_{rec}^{−γ}$ misst?

##### Lösung Aufgabe 6:

Für einfacheres Typesetting habe ich $E_\mathrm{real}=E$ und $E_\mathrm{rec}=E'$ gesetzt.


$$ \phi(E') = \phi_0 E'\,^{-\gamma} \qquad P(E) = \left|\frac{\mathrm{d}\frac{E}{E'}}{\mathrm{d}E}\right|P(\frac{E}{E'}) = \frac{1}{E'}\;\mathcal{L}\mathcal{N}(\mu,\sigma^2) 
\qquad \phi(E) = \int\limits_{0}^{\infty} \mathrm{d}E'\,P(E)\;\phi(E')$$ 

Mit der logarithmischen Normalverteilung $\mathcal{L}\mathcal{N}(\mu,\sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} \frac{E'}{E} \exp\left(-\frac{1}{2\sigma^2}(\log\frac{E}{E'}-\mu)^2\right)$. Es ergibt sich

$$ \phi(E) = E^{-(\gamma+1)} \frac{\phi_0}{\sqrt{2\pi\sigma^2}} \int\limits_{0}^{\infty} \mathrm{d}E'\, \exp\left(-\frac{1}{2\sigma^2}(\log\frac{E}{E'}-\mu)^2+\gamma\log\frac{E}{E'}\right)$$ 

Substituiere $X = \log\frac{E}{E'}-\mu \Longleftrightarrow \mathrm{dE'} = E\,\mathrm{dX}$. Es folgt

$$ 
\begin{aligned}
\phi(E) &= E^{-\gamma} \frac{\phi_0}{\sqrt{2\pi\sigma^2}}\;e^{\mu(\gamma-1)} \int\limits_{-\infty}^{\infty} \mathrm{d}E'\, \exp\;\left(\;-\,(\underbrace{\frac{1}{2\sigma^2}}_{=a}X^2+\underbrace{(1-\gamma)}_{=b}X\;)\right) \\
&= E^{-\gamma} \frac{\phi_0}{\sqrt{2\pi\sigma^2}}\;e^{\mu(\gamma-1)} \sqrt{\frac{\pi}{a}} \exp\;\left(\frac{b^2}{4a}\right) \\
&= E^{-\gamma}\phi_0\;e^{\mu(\gamma-1)}\exp\;\left(\frac{\sigma^2}{2}(1-\gamma)^2\right) \\
&= E^{-\gamma}\;\phi_0\;\exp\left(\frac{\sigma^2}{2}(1-\gamma)^2\right)
\end{aligned}
$$

Da $\;\langle\frac{E}{E'}\rangle\;=1$ und somit $\mu=\log(\;\langle\frac{E}{E'}\rangle\;)=0$ folgt.

Rotierende nichtstarre geladene Korper, wie z.B. Gase, in lokal gebunden Systemen erzeugen mitunter stabile Magnetfelder. Solche Magnetfelder sind u.a. wesentlich fur die Emission kosmischer Strahlung.

#### (a) Leiten Sie unter Annahme Newtonscher Gravitation einen Ausdruck fur die Kreisfrequenz, Umlaufzeit und die Geschwindigkeit eines Masseelements der Kreisscheibe her. 

##### Lösung Aufgabe 7a:

Mit $V_N = -\frac{M}{r}$ ergibt sich $F_N = -\nabla\,V_N = \frac{M}{r²}$. Angenommen das infinitesimale Massenelement bewegt sich auf einer Kreisbahn mit Radius $r$ erfährt es eine Zentrifugalkraft $F_Z = \frac{v²}{r}$. Durch gleichsetzten ergibt sich also

$$ \frac{M}{r²} = \frac{v²}{r} \Longleftrightarrow v = \sqrt{\frac{M}{r}}.$$

Die gesuchten Größen berechnen sich dann als

$$ f = \frac{v/r}{2\pi} = \frac{1}{2\pi} \sqrt{\frac{M}{r³}}, $$ 

$$ T = \frac{1}{f} = 2\pi \sqrt{\frac{r³}{M}}. $$

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [22, 12]
plt.rcParams.update({'font.size': 20})
plt.rc('axes', labelsize=26)
import numpy as np

# define the potential
M, r_max = 1, 30
theta, r = np.mgrid[slice(0, 2*np.pi, 0.05),slice(1, r_max, 0.05)]

fig = plt.figure()

for i, phi_dot in enumerate(np.linspace(-1.,1.,15),1):
    
    V_eff = np.array(-M/r * (1-phi_dot * np.sin(theta)**2))

    ax = fig.add_subplot(3,5,i,polar=True)
    ax.set_title(r"$\dot{\phi}$ = %.2f"%(phi_dot),pad=30)
    contour = ax.contourf(theta, r, V_eff, levels=30,cmap="gist_heat")
    ax.set_theta_zero_location("N")
    ax.set_rmax(r_max),ax.set_rmin(0)
    ax.set_xticklabels([])
    ax.set_yticklabels([])

plt.subplots_adjust(hspace=0.6, wspace=0.1)
plt.show()
```

#### (e) Stellen Sie das Magnetfeld, dass von geladenem Plasma in der Akkretionsscheibe des schwarzen Loches erzeugt wird, grafisch dar.

##### Einschub zur Berechnung von $I(r)$:

$$ I(r) = \frac{\Delta Q(r)}{t} = \frac{\rho\cdot A\cdot v(r)\cdot t}{t} \propto v(r)$$

Bestimme also $v(r)=\dot{\phi}(r)$ über Ansatz $F_Z = r^2\dot{\phi}^2 = \frac{M}{r^2} = -\nabla\;V_\mathrm{eff.} = F_G$. Es folgt

$$ \dot{\phi}_{1/2} = \frac{M}{2r} \pm \sqrt{\frac{M^2}{4r^{10}} + 4\frac{M}{r^3}}$$

mit $\dot{\phi}_1$ für prograde Orbits, $\dot{\phi}_2$ für retrograde Orbits, denke ich mal? Ich habe die Lösung mit dem Plus gewählt.

```python
from scipy.interpolate import griddata
from mpmath import ellipk as ell_1
from mpmath import ellipe as ell_2
from mpmath import ellippi as ell_3

# general formula and definitions
nPoints, nRings = 100, 50
r_min, r_max = -2000,2000
z_min, z_max = -1500,1500
l_disk, m_BH = 0.5,1
I_ring = lambda r: m_BH/(2*r) + np.sqrt(m_BH**2/(4*r**10) + 4*m_BH/r**3) #  - np.sqrt(m_BH/r) for Newtonian potential

# set up simulation canvas
r = np.array([np.linspace(r_min,r_max,nPoints) for i in range(nPoints)]).flatten()
z = np.array([[ele]*nPoints for ele in np.linspace(z_min,z_max,nPoints)]).flatten()

# returns the magnetic field B along given axis induced by a plasma disc of Radius R at point [rho,z]
# note that the problem is rotationally symmetric, so we don't really care about the phi component.
def B(R,rho,z,axis):
    
    # some prior definitions
    xi_min = z-l_disk/2
    xi_max = z+l_disk/2
    n = 4*R*rho/(rho+R)**2
    sign = [1.,-1.]
    B = 0
    
    # calculate the B field
    for l, xi in enumerate([xi_min,xi_max]):
        m = 4 * R*rho /((rho+R)**2 + xi**2)
    
        if axis=="r":
            B += sign[l] * (I_ring(R)/l_disk * 1/rho * ( np.sqrt( (rho+R)**2 + xi**2 ) * ( (2-m)*ell_1(m) - 2*ell_2(m) ) ))
        elif axis=="z":
            B += sign[l] * (2*I_ring(R)/l_disk*(xi/np.sqrt((rho+R)**2+xi**2)*((rho-R)/(rho+R)*ell_3(n,np.pi/2,m)-ell_1(m))))

    return B

def pseudo_streamplot(r,z,Br,Bz,density,exclude=[150,1100],alpha=0.1):
    
    # exclude central region arrows from blowing up the plot
    constraint_z = np.where(np.abs(z)<exclude[0])[0]
    constraint_r = np.where(np.abs(r)<exclude[1])[0]
    idx = [i for i in constraint_z if i in constraint_r]
    r_t = [r[i] for i in range(nPoints**2) if i not in idx]
    z_t = [z[i] for i in range(nPoints**2) if i not in idx]
    Br_t = [Br[i] for i in range(nPoints**2) if i not in idx]
    Bz_t = [Bz[i] for i in range(nPoints**2) if i not in idx]

    # draw the stream plot
    plt.quiver(r_t[::density],z_t[::density],Br_t[::density],Bz_t[::density],color="white",alpha=alpha)
    
def heatmap(x,z,Br,Bz,density):
    grid_r,grid_z = np.mgrid[-2000:2000:200j,-2000:2000:200j]
    heat_data = np.sqrt(Br**2 + Bz**2)
    ref_points = [[r[i],z[i]] for i in range(len(heat_data))]
    heat_data_interpolate = griddata(ref_points,heat_data,(grid_r,grid_z),method="cubic")
    
    plot = plt.imshow(heat_data_interpolate.T,extent=(-2000,2000,-2000,2000),cmap="plasma",aspect="auto")
    plt.colorbar(plot,label=r"|$\vec{B}(r,z)$|")
```

```python
from IPython.display import clear_output

B_r, B_z = np.zeros_like(r), np.zeros_like(z)

# Calculate the resulting B field
for i,R in enumerate(np.linspace(1,1000,nRings),1):
    for j in range(nPoints**2):
        B_r[j] += B(R,r[j],z[j],"r")
        B_z[j] += B(R,r[j],z[j],"z")
        
        print(f"Calculating B-Field for Ring {i}/{nRings}: {1e2*j/nPoints**2:.1f}%")
        clear_output(wait=True)
```

```python
# plotting the resultant magnetic field
pseudo_streamplot(r,z,B_r,B_z,density=3,exclude=[200,400],alpha=0.15)
heatmap(r,z,B_r,B_z,density=3)
plt.ylim(-1500,1500)
plt.xlim(-2000,1950)
plt.title(f"m_BH = {m_BH}: B-Field for {nRings} Rings with l_disk = {l_disk} r_s",pad=40)
plt.xticks(np.arange(-1500,1501,500),np.abs(np.arange(-1500,1501,500)))
plt.xlabel(r"$r\;(r_{s})$")
plt.ylabel(r"$z\;(r_{s})$")
plt.savefig("BH_magnetic_field.png")
plt.show()
```
